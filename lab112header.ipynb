```python
DATA = pd.read_csv("spam.csv", encoding = 'latin-1', usecols = (0,1)) # читаємо дані з файлу spam.csv
        
data = DATA.to_numpy() # перетворюємо Pandas DataFrame в Numpy Array

pd.options.display.max_colwidth = 1000
DATA # проглядаємо дані

# pandas тут використано тільки щоб прочитати дані і відобразити їх у вигляді гарної табличики
# далі працюємо зі змінною data, яка має тип "numpy array", як і в пункті 11.1
```

```python
# утворюємо два одновимірні масиви

h = data[data[:,0] == 'ham',:] 
H = h[:,1] # одновимірний масив, елементами якого є рядки (string) - ham-повідомлення 

s = data[data[:,0] == 'spam',:]
S = s[:,1] # одновимірний масив, елементами якого є рядки (string) - spam-повідомлення

print(len(data)) #число всіх повідомлень в spam.csv
print(len(H)) # число ham-повідомлень
print(len(S)) # число spam-повідомлень

# далі виконуємо роботу згідно завдання
```
